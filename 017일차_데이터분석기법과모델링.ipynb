````python
#  : 0,  : 1
y =  (df['CO(GT)'] >= 4).astype(int)
````

````python
df.shape, y.shape
````

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df,y,test_size=0.2, random_state=42)
````

````python
#   
print(y_train.value_counts())
import seaborn as sns
y_train.value_counts().plot(kind='bar')
````

````python
# 
from imblearn.over_sampling import RandomOverSampler
model_base =  RandomOverSampler(random_state=42)
x_train_over, y_train_over =  model_base.fit_resample(X_train, y_train)
print(f'  : {y_train.value_counts()}')
print(f'  : {y_train_over.value_counts()}')
````

````python
# 571 - 90 = 481
x_train_over[y_train_over ==1].duplicated().sum()
````

````python
# 

from imblearn.under_sampling import RandomUnderSampler
model_base =  RandomUnderSampler(random_state=42)
x_train_under, y_train_under =  model_base.fit_resample(X_train, y_train)
print(f'  : {y_train.value_counts()}')
print(f'  : {y_train_under.value_counts()}')
````

````python
# SMOTE
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
droped_x_train =  X_train.drop(['Date','Time'], axis = 1)
x_train_smote, y_train_smote =  smote.fit_resample(droped_x_train, y_train)

print(f'SMOTE  : {y_train.value_counts()}')
print(f'SMOTE  : {y_train_smote.value_counts()}')
print(f' : {x_train_smote.duplicated().sum()}')
````

````python
import numpy as np
import random
import missingno as msno
nan_sample = X.replace(-200, np.nan)  # -200   nan replace
nan_sample.dropna(inplace = True)  #  
# CO(GT)     np.nan 
nan_sample.reset_index(inplace=True)  #   ( )

# 10 ~ 100 
# 150 ~ 800   100
original_data = nan_sample.copy()
nan_index = list(random.sample(range(150,801),100) + list(range(10,101))  )
nan_sample.loc[nan_index , 'CO(GT)'] = np.nan
msno.matrix(nan_sample)
````

````python
interpolated_data = {}
# print(nan_index)
#   
df_linear =  nan_sample['CO(GT)'].interpolate(method = 'linear')
interpolated_data[''] = df_linear

# 
from sklearn.metrics import mean_squared_error
# 
real = original_data.loc[nan_index, 'CO(GT)'].to_numpy()
# 
predict = interpolated_data[''][nan_index].to_numpy()
# 
from sklearn.metrics import mean_squared_error
mean_squared_error(real,predict)  # 2.154003054101222
````

````python
interpolated_name = [ 'linear',  'polynomial' ,'spline', 'nearest' ,'akima','pchip' ]
interpolated_data = {}

real = original_data.loc[nan_index, 'CO(GT)'].to_numpy()

for bokan in interpolated_name:
  if bokan == 'polynomial':
    interpolated_data[bokan] = nan_sample['CO(GT)'].interpolate(method=bokan, order=2)
  elif bokan in 'spline':
    interpolated_data[bokan] = nan_sample['CO(GT)'].interpolate(method=bokan, order=3)
  else:
    interpolated_data[bokan] = nan_sample['CO(GT)'].interpolate(method=bokan)

for bokan in interpolated_name:
  predict = interpolated_data[bokan][nan_index].to_numpy()
  print(f'{bokan} : {mean_squared_error(real,predict)}')
````
