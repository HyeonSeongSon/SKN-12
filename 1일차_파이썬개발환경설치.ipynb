변수의 선언과 할당
  - 파이썬에서는 =(대입 연산자)를 사용하여 변수를 선언하고 값을 할당합니다.

```
x = 10  # x라는 변수에 정수 10을 저장
name = "Python"  # name 변수에 문자열 "Python" 저장
pi = 3.14  # pi 변수에 실수 3.14 저장
```
  - 파이썬은 동적 타이핑(dynamic typing) 언어
    - 변수 선언 시 자료형을 따로 지정하지 않아도 됨
    - 변수에 저장된 값을 기준으로 자동으로 타입이 결정됨

변수의 특징
  - 대소문자 구분
    - 변수명은 대소문자를 구분합니다.

```

print(age)  # 출력: 25

```
  - 유효한 변수명 규칙

    - 숫자로 시작할 수 없음
    - 파이썬의 예약어(keyword)는 사용할 수 없음

**자료형**
  - 정수 (int): 정수 값을 저장합니다.
    - 예를 들어, x = 10과 같이 정수를 변수에 할당할 수 있습니다.
  - 실수 (float): 소수점 이하 자릿수를 갖는 값을 저장합니다.
    - 예를 들어, y = 3.14와 같이 실수를 변수에 할당할 수 있습니다.
  - 문자열 (str): 텍스트 데이터를 저장합니다.
    - 따옴표(' 또는 " )로 묶어서 문자열을 만들 수 있습니다. 예를 들어, name = "Alice"와 같이 문자열을 변수에 할당할 수 있습니다.
  - 불리언 (bool): 참(True) 또는 거짓(False) 값을 저장합니다.
    - 예를 들어, is_valid = True와 같이 불리언 값을 변수에 할당할 수 있습니다.
  - 리스트 (list): 여러 개의 값을 순서대로 저장하는 데 사용됩니다.
    - 대괄호([])를 사용하여 리스트를 만들 수 있습니다. 예를 들어, numbers = [1, 2, 3, 4, 5]와 같이 리스트를 변수에 할당할 수 있습니다.
  - 튜플 (tuple): 리스트와 유사하지만 한 번 정의되면 값을 변경할 수 없는 불변의 데이터 구조입니다. 괄호 ()를 사용하여 튜플을 만들 수 있습니다.
    - 예를 들어 point = (10, 20)과 같이 튜플을 변수에 할당할 수 있습니다.
  - 딕셔너리 (dict): 키-값 쌍으로 데이터를 저장하는 데 사용됩니다. 중괄호({})와 콜론(:)을 사용하여 딕셔너리를 만들 수 있습니다.
    - 예를 들어, person = {"name": "Bob", "age": 30}과 같이 딕셔너리를 변수에 할당할 수 있습니다.
  - 파이썬은 동적 타입 언어이므로 변수를 선언할 때 타입을 명시적으로 지정하지 않습니다. 변수에 값을 할당하면 파이썬 인터프리터가 자동으로 데이터 타입을 추론합니다. 그러나 사용하려는 데이터의 타입을 항상 염두에 두는 것이 좋습니다.


**조건문**
  - 조건문(Conditional Statement)은 특정 조건을 만족할 때만 코드가 실행되도록 제어하는 구문. 즉, 프로그램이 특정 조건을 검사하고 그 결과(True 또는 False)에 따라 다른 동작을 수행할 수 있도록 함.
```
if 조건:
    실행할 코드
elif 다른_조건:
    실행할 코드
else:
    실행할 코드
```
  - if: 주어진 조건이 True일 때 실행
  - elif (선택 사항): 앞의 if 조건이 False이고, 새로운 조건이 True이면 실행
  - else (선택 사항): 모든 조건이 False일 때 실행

if 문 (기본 조건문)
  - if 문을 사용하면 특정 조건이 참일 때만 코드가 실행됩니다.
```
age = 20
if age >= 18:
      print("성인입니다.")  # 출력: 성인입니다.
```

if-else 문
  - 조건이 참이면 if 블록 실행, 거짓이면 else 블록 실행
```
age = 15
if age >= 18:
    print("성인입니다.")
else:
    print("미성년자입니다.")  # 출력: 미성년자입니다.
```

if-elif-else 문 (여러 개의 조건)
  - 여러 조건을 검사할 때 elif(else if)를 사용할 수 있습니다.
```
score = 85
if score >= 90:
    print("A 학점")
elif score >= 80:
    print("B 학점")  # 출력: B 학점
elif score >= 70:
    print("C 학점")
else:
    print("F 학점")
```






순환문
  - 순환문(또는 반복문)은 특정 코드 블록을 여러 번 실행할 때 사용하는 문법
  - 순환문 사용 이유
    - 코드 반복을 줄여서 효율적
    - 같은 작업을 여러 번 수행할 때 유용
    - 대량의 데이터를 다룰 때 필수적

  - for문
    - 리스트, 튜플, 문자열 등 순차적인 자료형을 순회할 때 사용

        `fruits = ["사과", "바나나", "포도"]`
        ```
      for fruit in fruits:
        print(fruit)
      ```
  - while문
    - 조건이 참(True)인 동안 계속 반복
      ```
      count = 1
      while count <= 5:
          print(f"반복 횟수: {count}")
          count += 1
      ```







함수
  - 함수는 어떤 작업을 수행하는 코드의 묶음. 같은 코드를 반복해서 작성하지 않고, 필요할 때마다 호출해서 사용
  - 함수 사용 이유
    - 코드를 재사용할 수 있어서 효율적
    - 코드가 짧고 가독성이 좋아짐
    - 유지보수와 수정이 쉬워짐
```
def 함수이름(매개변수1, 매개변수2, ...):
    """함수 설명 (docstring)"""
    실행할 코드
    return 반환값  # 선택 사항
```
```
# return을 사용하여 함수를 실행한 결과를 반환 할 수 있음
def add(a, b):
    return a + b

      print(result)  # 출력: 8
```
