# 셀레니움 + 뷰티풀

```python
# selenium
from selenium import webdriver
from selenium.webdriver.common.by import By

# beautifulsoup
from bs4 import BeautifulSoup

import time

url = 'http://www.naver.com/'

driver = webdriver.Chrome()
driver.get(url)

time.sleep(1)
button = driver.find_element(By.XPATH, '//*[@id="shortcutArea"]/ul/li[5]/a/span[1]')
button.click()
time.sleep(1)

# 뉴스를 클릭하면 새로운 탭이 생겨서 새로운 탭으로 경로를 바꿔줘야함
driver.switch_to.window(driver.window_handles[1])
time.sleep(2)

# 현재 페이지소스를 뷰티플 숍으로 전달
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
for news_title in soup.select('strong.cnf_news_title'):
    print(news_title.text)
driver.quit()
```

# 동적으로 생성되는 컨텐츠 데이터 가져오기

```python
# selenium
from selenium import webdriver
from selenium.webdriver.common.by import By

# beautifulsoup
from bs4 import BeautifulSoup

import time

url = 'https://www.coffeebeankorea.com/store/store.asp'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(1)

driver.execute_script("storePop2('400')")
time.sleep(1)
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
tr_lists = soup.select('#matizCoverLayer0Content > div > div > div.store_txt > table > tbody:nth-child(1) > tr')
names = ['매장명','영업시간','주차','주소','전화번호','기타']
saved_dict = {'매장명': soup.select('div.store_txt > h2')[0].text}

for idx,tr in enumerate(tr_lists):
    saved_dict[names[idx]] = tr.find('td').text
print(saved_dict)


driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
from bs4 import BeautifulSoup
import time
import requests

url = 'https://www.index.go.kr/unity/potal/main/EachDtlPageDetail.do?idx_cd=1257'
driver = webdriver.Chrome()
driver.get(url)
time.sleep(1)

iframe = driver.find_element(By.ID, 'showStblGams')
driver.switch_to.frame(iframe)  # iframe으로 변환
# iframe(Inline Frame)은 웹페이지 안에 또 다른 HTML 문서를 삽입하는 태그. Selenium은 기본적으로 iframe 외부에서 동작하기 때문에, iframe 내부의 요소를 찾으려면 switch_to.frame()을 사용해야함.

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')   # 소스 파싱
for tr in soup.select('#t_Table_125701 > tbody > tr'):
    print(tr.text)
    print('-'*50)
driver.quit()
```

```python
def data_parser(driver,page_num):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    div_lists = soup.select(f'#main_pack > div.sc_new.cs_common_module.case_list.color_5._cs_car_month_list > div.cm_content_wrap > div.cm_content_area > div > div._car_panel_wrapper > div:nth-child({page_num}) > div > div')
    for div in div_lists:
        print(div.find('strong',class_='title').text ,
              div.find_all('span',class_='info_txt')[0].text,
              div.find_all('span',class_='info_txt')[1].text
             )
```
